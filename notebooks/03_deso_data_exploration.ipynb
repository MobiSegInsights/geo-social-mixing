# DeSO Data Exploration

Explore the structure of DeSO data files before harmonization.

```python
import pandas as pd
import geopandas as gpd
from pathlib import Path

DESO_DIR = Path('/workspace/dbs/deso')
print(f"DeSO directory: {DESO_DIR}")
print(f"Files: {list(DESO_DIR.glob('*'))}")
```

## 1. Geometry File

```python
# Load geometry
gdf = gpd.read_file(DESO_DIR / 'DeSO_2025.gpkg')

print(f"Shape: {gdf.shape}")
print(f"\nColumns: {gdf.columns.tolist()}")
print(f"\nDtypes:\n{gdf.dtypes}")
print(f"\nCRS: {gdf.crs}")
print(f"\nFirst 5 rows:")
gdf.head()
```

```python
# Check which column contains DeSO codes
print("Sample values from each column:")
for col in gdf.columns:
    if col != 'geometry':
        print(f"\n{col}:")
        print(gdf[col].head(3).tolist())
```

## 2. Birth Background File

```python
# Load birth background - check both with and without skiprows
print("=== First row (header description) ===")
with open(DESO_DIR / 'deso_birth_background_gender.csv', 'r', encoding='latin1') as f:
    print(f.readline())
    print(f.readline())
    print(f.readline())

print("\n=== Loading with skiprows=1 ===")
df_birth = pd.read_csv(DESO_DIR / 'deso_birth_background_gender.csv', 
                       skiprows=1, encoding='latin1')
print(f"Shape: {df_birth.shape}")
print(f"\nColumns: {df_birth.columns.tolist()}")
print(f"\nFirst 10 rows:")
df_birth.head(10)
```

```python
# Check unique values
print("Unique values in column 1 (birth region):")
print(df_birth.iloc[:, 1].unique())

print("\nUnique values in column 2 (gender):")
print(df_birth.iloc[:, 2].unique())

print("\nSample DeSO codes from column 0:")
print(df_birth.iloc[:, 0].head(10).tolist())
```

```python
# Filter to 'totalt' gender
df_birth.columns = ['region', 'birth_region', 'gender', 'count']
df_birth_total = df_birth[df_birth['gender'] == 'totalt'].copy()
df_birth_total = df_birth_total[df_birth_total['birth_region'] != 'totalt'].copy()

print(f"After filtering to 'totalt' gender (excl. 'totalt' birth region): {len(df_birth_total)} rows")
print(f"\nUnique birth regions: {df_birth_total['birth_region'].unique()}")
print(f"\nSample:")
df_birth_total.head(10)
```

```python
# Pivot
birth_wide = df_birth_total.pivot(index='region', columns='birth_region', values='count').reset_index()
print(f"Pivoted shape: {birth_wide.shape}")
print(f"\nColumns: {birth_wide.columns.tolist()}")
birth_wide.head()
```

## 3. Income File

```python
print("=== First row (header description) ===")
with open(DESO_DIR / 'deso_income_gender.csv', 'r', encoding='latin1') as f:
    print(f.readline())
    print(f.readline())

print("\n=== Loading with skiprows=1 ===")
df_income = pd.read_csv(DESO_DIR / 'deso_income_gender.csv', 
                        skiprows=1, encoding='latin1')
print(f"Shape: {df_income.shape}")
print(f"\nColumns: {df_income.columns.tolist()}")
print(f"\nFirst 5 rows:")
df_income.head()
```

```python
# Check unique values
print("Unique income types (column 1):")
print(df_income.iloc[:, 1].unique())

print("\nUnique genders (column 2):")
print(df_income.iloc[:, 2].unique())
```

```python
# Filter to 'sammanräknad' and 'totalt'
df_income.columns = ['region', 'income_type', 'gender', 'q1_pct', 'q2_pct', 
                     'q3_pct', 'q4_pct', 'median_tkr', 'mean_tkr', 'n_persons']

df_income_filt = df_income[
    (df_income['income_type'].str.contains('sammanr', na=False)) &
    (df_income['gender'] == 'totalt')
].copy()

print(f"After filtering: {len(df_income_filt)} rows")
print(f"\nSample:")
df_income_filt.head()
```

## 4. Cars File

```python
print("=== First row (header description) ===")
with open(DESO_DIR / 'deso_passenger_cars_in_traffic.csv', 'r', encoding='latin1') as f:
    print(f.readline())
    print(f.readline())

print("\n=== Loading with skiprows=1 ===")
df_cars = pd.read_csv(DESO_DIR / 'deso_passenger_cars_in_traffic.csv',
                      skiprows=1, encoding='latin1')
print(f"Shape: {df_cars.shape}")
print(f"\nColumns: {df_cars.columns.tolist()}")
print(f"\nFirst 10 rows:")
df_cars.head(10)
```

## 5. Check Join Keys

```python
# Compare DeSO codes across files
print("DeSO code samples from each file:")
print("\nGeometry file:")
if 'deso' in gdf.columns:
    print(f"  Column: 'deso'")
    print(f"  Sample: {gdf['deso'].head(5).tolist()}")
elif 'DeSO' in gdf.columns:
    print(f"  Column: 'DeSO'")
    print(f"  Sample: {gdf['DeSO'].head(5).tolist()}")
else:
    print(f"  Available columns: {[c for c in gdf.columns if c != 'geometry']}")
    for col in gdf.columns:
        if col != 'geometry':
            print(f"  {col}: {gdf[col].head(3).tolist()}")

print("\nBirth file (region column):")
print(f"  Sample: {birth_wide['region'].head(5).tolist()}")

print("\nIncome file (region column):")
print(f"  Sample: {df_income_filt['region'].head(5).tolist()}")

print("\nCars file:")
df_cars.columns = ['region', 'status', 'count']
print(f"  Sample: {df_cars['region'].head(5).tolist()}")
```

```python
# Check for matches
print("Checking overlap:")

# Find the DeSO code column in geometry
geo_deso_col = None
for col in ['deso', 'DeSO', 'desokod']:
    if col in gdf.columns:
        geo_deso_col = col
        break

if geo_deso_col:
    geo_codes = set(gdf[geo_deso_col])
    birth_codes = set(birth_wide['region'])
    income_codes = set(df_income_filt['region'])
    cars_codes = set(df_cars['region'])
    
    print(f"\nGeometry: {len(geo_codes):,} DeSO codes")
    print(f"Birth: {len(birth_codes):,} DeSO codes")
    print(f"Income: {len(income_codes):,} DeSO codes")
    print(f"Cars: {len(cars_codes):,} DeSO codes")
    
    print(f"\nOverlap with geometry:")
    print(f"  Birth: {len(geo_codes & birth_codes):,} / {len(birth_codes):,} ({len(geo_codes & birth_codes)/len(birth_codes)*100:.1f}%)")
    print(f"  Income: {len(geo_codes & income_codes):,} / {len(income_codes):,} ({len(geo_codes & income_codes)/len(income_codes)*100:.1f}%)")
    print(f"  Cars: {len(geo_codes & cars_codes):,} / {len(cars_codes):,} ({len(geo_codes & cars_codes)/len(cars_codes)*100:.1f}%)")
    
    # Show sample mismatches
    birth_only = birth_codes - geo_codes
    if birth_only:
        print(f"\nSample codes in Birth but not Geometry: {list(birth_only)[:5]}")
    
    geo_only = geo_codes - birth_codes
    if geo_only:
        print(f"Sample codes in Geometry but not Birth: {list(geo_only)[:5]}")
else:
    print("Could not find DeSO code column in geometry file!")
    print(f"Available columns: {gdf.columns.tolist()}")
```

## 6. Test Join

```python
# Try joining
if geo_deso_col:
    # Rename for consistency
    gdf_test = gdf.rename(columns={geo_deso_col: 'deso_kod'})
    birth_test = birth_wide.rename(columns={'region': 'deso_kod'})
    
    # Test join
    result = gdf_test.merge(birth_test, on='deso_kod', how='left')
    
    print(f"Join result shape: {result.shape}")
    print(f"\nColumns: {result.columns.tolist()}")
    print(f"\nFirst 5 rows:")
    display(result.head())
    
    print(f"\nMissing data after join:")
    for col in result.columns:
        if col not in ['deso_kod', 'geometry']:
            n_missing = result[col].isna().sum()
            print(f"  {col}: {n_missing:,} missing ({n_missing/len(result)*100:.1f}%)")
```
